```python
import sys
sys.path.insert(0, '..')

import openpmd_api as io
import numpy as np
import matplotlib.pyplot as plt

bp_file = 'test_folder/parthenon.sl.x1.0.5.00000.bp'

series = io.Series(bp_file, io.Access.read_only)
it = series.iterations[0]
mesh = it.meshes['cons_lvl0']

print('Available fields:', list(mesh))

comp = mesh['density']
data = comp.load_chunk()
series.flush()

dx = mesh.grid_spacing
x0 = mesh.grid_global_offset
extent = [x0[1], x0[1] + dx[1]*data.shape[1],
          x0[0], x0[0] + dx[0]*data.shape[0]]

fig, ax = plt.subplots(figsize=(5, 4))
im = ax.imshow(data, origin='lower', extent=extent, cmap='viridis')
fig.colorbar(im, ax=ax, label='density')
ax.set_xlabel('x2')
ax.set_ylabel('x3')
ax.set_title('density  |  t = 0')
plt.show()

series.close()
```
